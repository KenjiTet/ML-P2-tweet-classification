```python
import pandas as pd
import csv
from keras.preprocessing.text import Tokenizer
from nltk.tokenize import word_tokenize
import os
import sys


# Now you can import from 'utils
```

```python
import gensim 
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
df_full = pd.read_pickle("tweets.pkl")


```

```python
df_full["tweet"][0]
```

```python
#work with the full or the small dataset depending on the uncommented line below
l_tweets = list(df_full["tweet"].apply(lambda x : word_tokenize(x))) #tokenize the tweets to get a list of the list of the words of each tweet
#l_tweets = list(df_small.apply(lambda x : word_tokenize(x))) #tokenize the tweets
```

## Train the word2vec model with gensim

```python
model = gensim.models.Word2Vec (l_tweets, vector_size=200, window=20, min_count=2, workers=10)
model.train(l_tweets,total_examples=len(l_tweets),epochs=10)
```

```python
word_vectors = model.wv #get only the word embedding vectors from the model
word_vectors.save_word2vec_format('w2v_full_w20_min2.txt', binary=False) #export the word embeddings
```
